```python
import pandas as pd
matches = pd.read_csv("../ipl-matches-study/matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

## find hightest run in perticular season

```python
matches.head(1)
```

```python
deliveries.head(1)
```

```python
data = deliveries.merge(matches, how="inner", left_on="match_id", right_on="id")
data
```

```python
group = data.groupby(["season", "batsman"])["batsman_runs"].sum().reset_index() # group on season and batsman
group.sort_values("batsman_runs", ascending=False).drop_duplicates(subset="season", keep="first").sort_values("season") # get heigetst run and remove duplicate seasons
```
