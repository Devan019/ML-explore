```python
import pandas as pd
data = pd.read_csv("deliveries.csv")
data.head(3)
```

```python
data.info()
```

find the most destructive death over bastman in history of ipl
- based on strike rate = (no of runs/ no of balls)*100
- also min balls are 200 in over 16-20

```python
death_bastmans = data[data["over"] > 15]
all_bastmans = death_bastmans.groupby("batsman")["batsman_runs"].count()
check = all_bastmans > 200
bastmans = all_bastmans[check].index.to_list()
bastmans
```

```python
final  =  data[data["batsman"].isin(bastmans)]
death_final = final["over"] > 15
final_group = final[death_final].groupby("batsman")

runs = final_group["batsman_runs"].sum()
balls = final_group["batsman_runs"].count()

strike_rate = (runs/balls)*100
strike_rate.sort_values(ascending=False).head()

```

```python
#new dataframe

death_stats = pd.DataFrame({
    "runs": runs,
    "balls":balls,
    "strike rate" : strike_rate
})
death_stats.sort_values( "strike rate" ,ascending=False)
```
