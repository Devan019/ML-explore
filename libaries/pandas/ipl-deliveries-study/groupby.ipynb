```python
import pandas as pd
data = pd.read_csv("deliveries.csv")
data
```

groupby - make a group of a common filed like a bastman

```python
bastsman = data.groupby("batsman")
bastsman.get_group("DA Warner").shape[0] # total bow played by da waner
```

get top-5 batsman which have most runes

```python
bastsman["batsman_runs"].sum().sort_values(ascending=False).head()
```

top 5 batsman who hit most 6's

```python
mask = data["batsman_runs"] == 6
data[mask].groupby("batsman")["batsman_runs"].count().sort_values(ascending=False).head()
```

a player's most run against a single team

```python
def getAgainstTeam(palyer : str):
  mask = data["batsman"] == palyer
  return data[mask].groupby("bowling_team")["batsman_runs"].sum().idxmax()

getAgainstTeam("V Kohli")
```
