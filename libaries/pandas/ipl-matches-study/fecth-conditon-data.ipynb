### Put condition on data

```python
data['toss_decision'] == "bat" # boolean series
```

```python
mask = data['toss_decision'] == "bat"
data[mask] # only get data of who win tossed and take bat
```

```python
def getCountOfTeamWin(team:str):
  mask = data['winner'] == team
  return data[mask].shape[0]

getCountOfTeamWin('Kings XI Punjab')
```

function - match after 2012 and play in hyderabad

```python
mask1 = data['date'] >= "2012-01-01"
mask2 = data['city'] == "Hyderabad"
data[mask1 & mask2].shape[0]
```
