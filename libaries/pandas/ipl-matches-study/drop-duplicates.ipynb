```python
import pandas as pd
data = pd.read_csv("matches.csv")
data
```

drop dublicate remove duplicates - ex remove duplicate city

```python
data.drop_duplicates(subset=["city"]).sort_values("city").head()
```

keep parameter for keep last or first occurence

```python
data.drop_duplicates(subset=["city"],keep="last").sort_values("city").head()
```

apply duplication on different cols

```python
data.drop_duplicates(subset=["city", "date"]) #city and date are same then remove it
```

find final winner of every season

```python
data.drop_duplicates(subset=["season"], keep="last").sort_values("season")[["season", "winner"]]
```
